```python
from urllib.request import urlopen
import json
import pandas as pd
import plotly.express as px

with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)

pairs = {"Cases" : [df5, 'fips_x', 'cases', 'Inferno_r', 'month', 'cases', '# Cases'],
         "Deaths" : [df5, 'fips_x', 'deaths', 'Inferno_r', 'month', 'deaths', '# Fatalities'],
         "Device Count" : [df1, 'county', 'device_count', 'ice_r','Month',
                                       'device_count', '# Devices'],
         "% Time in Quarantine" : [df1, 'county', 'median_percentage_time_home', 'Viridis','Month',
                                       'median_percentage_time_home', '% Time in Quarantine'],
         "Travel Frequency" : [df1, 'county', 'distance_traveled_from_home', 'BuPu','Month',
                                       'distance_traveled_from_home', 'Distance From Home'],
         
        }
```

```python
df1.dropna().head()
```

```python
def mapping(user_entry):
    values = pairs[user_entry]
    range_max = int(list(values[0][values[2]].quantile([.98]))[0])
    fig = px.choropleth_mapbox(values[0].dropna(), geojson=counties, locations=values[1], color=values[2],
                           color_continuous_scale=values[3],
                           range_color=(0, range_max),
                           mapbox_style="carto-positron",
                           zoom=3, center = {"lat": 37.0902, "lon": -95.7129},
                           opacity=0.5,
                           animation_frame = values[4],
                           animation_group = values[5],
                           labels={values[5]:values[6]}
                          )
        
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
```
